```python
base_dir = "/Users/lzj/data/QL/"
```

```python
%%time
qlcp21a.do_all(
    ini_file="xl60.ini", 
    raw_dir=base_dir+"raw/20210404zqy/",
    lst_dir=base_dir+"lst/20210404_60/", 
    red_dir=base_dir+"red/20210404_60/",
    ct_coord=("15:24:07", "+36:52:00"),
    steps="lbfipocdg",
)
```

上面是自动选出了若干颗参考星，也正确地从图中找到了我们的目标星。但是从图中可以看到，在R波段由于1号参考星在某一幅图像中测光结果差异太大，导致R波段彻底跑偏。经过检查数据发现在R_102图像中1号参考星附近出现异常星。

我们试着把`cali_dis_limit`参数调小，从默认的10.0改为5.0，检查效果。

另外，因为已知目标，并且列表已经生成了，所以这里直接通过参数指定目标，并且只重做最后kcdg步骤（k是隐含的）。因为是覆盖原结果，所以必须加上overwrite参数

```python
%%time
qlcp21a.do_all(
    ini_file="xl60.ini", 
    obj="YZBoo", band="BVR",
    raw_dir=base_dir+"raw/20210404zqy/",
    lst_dir=base_dir+"lst/20210404_60/", 
    red_dir=base_dir+"red/20210404_60/",
    ct_coord=("15:24:07", "+36:52:00"),
    steps="cdg",
    extra_config={"cali_dis_limit": 5.0},
    overwrite=True
)
```

注意，上面的输出中含有大量的警告信息，有些是因为fits图像不够规范造成的，有些是因为matplotlib升级造成的，还有一些就是由于时间无法更新造成的。
